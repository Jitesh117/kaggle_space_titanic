```python
numerical_cols = df.select_dtypes(include=['int', 'float']).columns
categorical_cols = df.select_dtypes(include='object').columns
```

```python
df['VIP'].unique()
```

```python
for category in numerical_cols:
    df[category].fillna(value=df[category].mean(), inplace=True)

for category in categorical_cols:
    df[category].fillna(value=df[category].mode().iloc[0], inplace=True)
```

```python
df.head()
```

```python
# check the number of unique categorical values, to see if I should one hot encode them or go with some other approach

for category in categorical_cols:
    print(category, ":" ,len(df[category].unique()))
```

```python
df[['Deck', 'Num', 'Side']] = df['Cabin'].str.split('/', expand=True)
```

```python
df_bk = df.copy()
df.head()
```

```python
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols
```

```python
# One Hot encoding for categorical columns
# import necessary packages
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
encoder = LabelEncoder()
for category in categorical_cols:
    df[category] = encoder.fit_transform(df[category])
```

```python
df.head()
```

```python
input_df = df.drop(columns=['Transported', 'PassengerId'])
target_df = df['Transported']
```

```python
target_df.head()
```

```python
# train terst split
from sklearn.model_selection import train_test_split
train_val_inputs, test_inputs = train_test_split(input_df, test_size=0.2, random_state=42)
train_inputs, val_inputs = train_test_split(train_val_inputs, test_size=0.25, random_state=42)

train_val_target, test_target = train_test_split(target_df, test_size=0.2, random_state=42)
train_target, val_target = train_test_split(train_val_target, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC, NuSVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import RidgeClassifier, PassiveAggressiveClassifier, Perceptron
from sklearn.multioutput import MultiOutputClassifier

# Experimental Classifiers
from xgboost import XGBClassifier  # Assuming you have XGBoost installed separately
from sklearn.neighbors import NearestCentroid
```

```python
models = [
    LogisticRegression(),
    SGDClassifier(),
    SVC(),
    NuSVC(),
    LinearSVC(),
    KNeighborsClassifier(),
    GaussianNB(),
    MultinomialNB(),
    BernoulliNB(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    AdaBoostClassifier(),
    BaggingClassifier(),
    MLPClassifier(),
    RidgeClassifier(),
    PassiveAggressiveClassifier(),
    Perceptron(),
    # MultiOutputClassifier(RandomForestClassifier()),  # Example of MultiOutputClassifier
    XGBClassifier(),
    NearestCentroid(),
]
```

```python
xg_model = XGBClassifier()
xg_model.fit(train_inputs, train_target)
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(xg_model, file)
```

```python
with open('model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

loaded_model.score(val_inputs, val_target)
```

```python
for model in models:
    model.fit(train_inputs, train_target)
    print(model, ":" , model.score(val_inputs, val_target,))
    print(model, ":" , model.score(test_inputs, test_target,))
```

```python
from sklearn.ensemble import VotingClassifier

named_models = [('model{}'.format(i), model) for i, model in enumerate(models)]
voting_classifier = VotingClassifier(estimators=named_models, voting='hard')
voting_classifier.fit(train_inputs, train_target)

predictions = voting_classifier.predict(val_inputs)

# Evaluate the ensemble model
print('Ensemble Model Accuracy:', voting_classifier.score(val_inputs, val_target))
```
